# Brute-Force Password Retriever
In the following cell, I ask to the user to insert a list of characters and then a password. <br>
The algorithm is able to brute-force such password.

```python
# Import library for generate passwords
import itertools

# Ask user the list of characters
characters = str(input(
    """Insert the valid characters, following the seguent rule:
    - The list of character must be separe by a , and without any space
    Example: a,b,c,A,@,+,?,!\n"""))

# Now convert such string into a list of characters
characters = characters.replace(" ","")
list_c = characters.split(",")

# Remove from the list the invalid elements
for element in list_c:
    if len(element) > 1:
        list_c.remove(element)
list_c = [element for element in list_c if element != ''] # Empty element
list_c = list(set(list_c)) # Remove duplicates, the order doesn't matter

# Ask password with such characters
while True:
    password = str(input("Now insert a 1 word password with such characters: "))
    # Check the input
    not_in_list = False
    for c in password:
        if c not in list_c: # A character is not in the list of valid characters
            not_in_list = True
            break
    if not not_in_list: # All the characters are in the list of valid characters because not_in_list is still False
        break
        
# Now generate passwords with such characters until I found the correct one
password_found = False
current_length = 1
while not password_found:
    # Generate all the possible combinations with current_length characters
    for combo in itertools.product(list_c, repeat = current_length):
        generated_password = ''.join(combo)
        if generated_password == password:
            password_found = True
            break
    current_length += 1

print(f"I have brute-forced your password: {generated_password}")
```

# Caesar's Cipher Resolution
In the following cell, I ask to the user to insert a message and the value of the key K used in the Caesar Cipher. <br>
Then the algorithm is able to retrive such encrypted message.

```python
# Define a function which receives as parameters the message and the key
# Return the encrypted message
def encrypt_caesar(message, key):
    encrypted_message = ""

    # Iterate through every character of the message
    for character in message:
        # Encrypt only alphabetic characters
        if character.isalpha():
            # Check if the character is uppercase or lowercase
            offset = 65 if character.isupper() else 97

            # Now use the key for encrypt the message
            encrypted_message += chr((ord(character) - offset + key) % 26 + offset)
        else: # Keep the other characters unchanged
            encrypted_message += character

    return encrypted_message

# Define a function which receives as parameters the encrypted message 
# Return all the possible pairs {key - decrypted message}
def decrypt_caesar_brute_force(encrypted_message):
    messages = {}
    
    # Try all possible shifts from 1 to 25
    for key in range(1, 26):
        decrypted_message = ""
        for character in encrypted_message:
            if character.isalpha():
                offset = 65 if character.isupper() else 97
                decrypted_message += chr((ord(character) - offset - key) % 26 + offset)
            else:
                decrypted_message += character
        messages[key] = decrypted_message  # Store results with the corresponding key
    
    return messages

# Ask user the message and the key
message = str(input("Insert a message: "))
key = int(input("Insert the value of the key K used in the Caesar Cipher: "))

# Encrypt the message
encrypted_message = encrypt_caesar(message, key)
print(f"\nEncrypted message: {encrypted_message}\n")

# Get all the possible decrypted messages
decrypted_messages = decrypt_caesar_brute_force(encrypted_message)

# Print all the decrypted messages
for x_key, y_message in decrypted_messages.items():
    print(f"Key = {x_key}: {y_message}")
    if y_message == message:
        correct_key = x_key
print(f"\nYour key is: {correct_key}")
```
